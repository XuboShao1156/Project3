```python
def vis(x) :
    x1 = x['CBR Rate'].values
    v1 = x['Variant 1'].values
    v2 = x['Variant 2'].values
    variant1 = str(v1[0])
    variant2 = str(v2[0])
    start1 = x['Start Time 1'].values
    start2 = x['Start Time 2'].values
    start3 = str(start1[0])
    start4 = str(start2[0])
    tp_1 = x['Throughput 1'].values
    tp_2 = x['Throughput 2'].values
    dr_1 = x['Drop Rate 1'].values
    dr_2 = x['Drop Rate 2'].values
    lt_1 = x['Latency 1'].values
    lt_2 = x['Latency 2'].values
    
    plt.figure(figsize=(8, 10))
    plt.subplot(3,1,1)
    plt.plot(x1, tp_1,label= "(variant1)" + variant1 + start3)
    plt.plot(x1, tp_2,label= "(variant2)" + variant2 + start4)
    
    plt.xlabel('CBR Rate(in Mbps)')
    plt.ylabel('Throughoutput(in Mbps)')
    leg = plt.legend(loc='upper right')
    plt.grid(True)
    plt.axis([0,11,0,10])
    
    
    # dp
    plt.subplot(3,1,2)
    plt.plot(x1, dr_1,label= "(variant1)" + variant1+ start3)
    plt.plot(x1, dr_2,label= "(variant2)" + variant2+ start4)
    
    plt.xlabel('CBR Rate(in Mbps)')
    plt.ylabel('Drop Rate(in Mbps)')
    leg = plt.legend(loc='upper right')
    plt.grid(True)
    plt.axis([0,11,0,0.6])
    
    # lt
    plt.subplot(3,1,3)
    plt.plot(x1, lt_1,label= "(variant1)" + variant1 + start3)
    plt.plot(x1, lt_2,label= "(variant2)" + variant2+ start4)
    
    plt.xlabel('CBR Rate(in Mbps)')
    plt.ylabel('Latency(in Mbps)')
    leg = plt.legend(loc='upper right')
    plt.grid(True)
    plt.axis([0,11,0,0.03])
    
    print(variant1)
```
